```python
data_wx['province'] = data_wx['area'].str.split("-", expand=True)[0]
data_wx['city'] = data_wx['area'].str.split("-", expand=True)[1]
# 针对四个直辖市北京、天津、上海、重庆将省份和城市赋值一致
data_wx.loc[data_wx['province'] == '北京', 'city'] = '北京'
data_wx.loc[data_wx['province'] == '上海', 'city'] = '上海'
data_wx.loc[data_wx['province'] == '天津', 'city'] = '天津'
data_wx.loc[data_wx['province'] == '重庆', 'city'] = '重庆'
```

**2. 针对学历、经验进行清理**

```python
data_wx['exp_level'].sample(5)
```

```python
data_wx['exp'] = data_wx['exp_level'].str.split("/", expand=True)[0]
data_wx['degree'] = data_wx['exp_level'].str.split("/", expand=True)[1]
```

```python
data_wx['degree'].unique()
```

```python
degree_map = {'大学本科': '本科', '大学本科以上': '本科', '不限': '学历不限', '大专以上': '大专', '大专': '大专',
              '硕士以上': '硕士', '不限以上': '学历不限', '硕士': '硕士', '博士以上': '博士', '中专': '中专',
              '高中以上': '高中', '职高': '中专'}

data_wx['degree'] = data_wx['degree'].map(degree_map)
```

```python
data_wx.sample(5)
```

**3. 针对工资进行清洗**
- 将工资转换为月平均工资

```python
data_wx['salary'].unique()
```

```python
# 定义函数求得每月的平均工资
def get_salary(data):
    paid_K = r"(.*)K-(.*)K"
    paid_W = r"(.*)W-(.*)W"
    paid = r"(.*)-(.*)/"
    if '面议' in data:
        return np.nan
    if '享公办教师薪资待遇' in data:
        return np.nan
    if 'K' in data and '月' in data:
        low, high = re.findall(pattern=paid_K, string=data)[0]
        return (float(low)+float(high))/2
    if 'W' in data and '年' in data:
        low, high = re.findall(pattern=paid_W, string=data)[0]
        return (float(low)+float(high))/2*10/12
    if 'K' not in data and '月' in data:
        low, high = re.findall(pattern=paid, string=data)[0]
        return (float(low)+float(high))/2/1000

data_wx['salary_clean'] = data_wx['salary'].apply(get_salary)
# 保留一位小数
data_wx['salary_clean'] = np.round(data_wx['salary_clean'], 1)
```

```python
data_wx.sample(5)
```

**4. 对教师授课层次进行清洗**
- 分为“幼儿园”、“小学”、“中学”、“其他”四类

```python
def get_type(data):
    if re.compile('[幼少早].').findall(data):
        return '幼儿园'
    elif re.compile('[小].').findall(data):
        return '小学'
    elif re.compile('[中].').findall(data):
        return '中学'
    else:
        return '其他'


data_wx['type'] = data_wx['title'].apply(get_type)
```

**2. 对 salary 进行清洗**

```python
data_jlc['salary'].unique()
```

```python
def get_salary_jlc(data):
    pat_jlc = r"(.*)K/MTH - (.*)K/MTH"
    if '00' in data:
        low, high = re.findall(pattern=pat_jlc, string=data)[0]
        return (float(low)+float(high))/2/1000
    else:
        low, high = re.findall(pattern=pat_jlc, string=data)[0]
        return (float(low)+float(high))/2

data_jlc['salary_clean'] = data_jlc['salary'].apply(get_salary_jlc)
```
